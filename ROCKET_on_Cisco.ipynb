### Creating functions to calculate execution time for each test case

```python
def splittint_time(time_data):
    
    time_data = time_data.split('T')
    date = time_data[0]
    time_data = time_data[1].split('+')
    h, m, s = time_data[0].split(':')
    return date, h, m, s
 
    
    
def time_in_sec(start_time, finish_time):
    sd, sh, sm, ss = splittint_time(start_time)
    fd, fh, fm, fs = splittint_time(finish_time)    
    
    if int(sh) > int(fh):
        finished_in_sec = ((int(fh)+24) * 3600 + int(fm) * 60 + int(fs))
        started_in_sec = (int(sh) * 3600 + int(sm) * 60 + int(ss))
        exec_time = finished_in_sec - started_in_sec
    else:
        finished_in_sec = (int(fh) * 3600 + int(fm) * 60 + int(fs))
        started_in_sec = (int(sh) * 3600 + int(sm) * 60 + int(ss))
        exec_time = finished_in_sec - started_in_sec
        
    return exec_time, (sh), sd
```

### Algorithm to calculate distance from current execution

```python
#function to reverse each row of the matrix
def reverse_row_element(array_data):
    list_data = array_data.tolist()
    list_data.reverse()
    return list_data

# defining failure matrix MF
MF = df_final.loc[:,'E1':'E3'].values

rows = len(MF)
columns = len(MF[0])

#calculating distance of failure for each test cases
distances_list = []

for i in range(rows):
    reversed_element = reverse_row_element(MF[i])
    if -1 in reversed_element:        
        for j in range(len(reversed_element)):
            if reversed_element[j] == -1:
                distances_list.insert(i, j+1)
                break
    else:
        #If distance == 0, means the test case has always been successful
        distances_list.insert(i, 0)
```

```python
#calculating number of change of state of Pass/Fail for each test cases
change_in_status = []

for i in range(rows):
    each_element = MF[i]
    counter = 0
    for j in range(len(each_element)-1):
        if each_element[j]!=each_element[j+1]:
            counter+=1
    change_in_status.append(counter)
```

```python
df_final['DIST']=distances_list
df_final['CHANGE_IN_STATUS']=change_in_status
```
